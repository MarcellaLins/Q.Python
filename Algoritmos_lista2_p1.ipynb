```python
#QUESTÃO 15

votos = [0, 0, 0, 0, 0, 0, 0]
total_votos = 0

print("0 - ENCERRAR \n1 - Windows Server \n2 - Unix \n3 - Linux \n4 - Netware \n5 - Mac OS \n6 - Outro")

voto = int(input("Digite o número correspondente à sua escolha: "))
while voto != 0:
    if 1 <= voto <= 6:
        votos[voto] += 1
        total_votos += 1
    else:
        print("Voto inválido! Por favor, digite um número entre 0 e 6.")
    voto = int(input("Digite o número correspondente à sua escolha: "))

percentuais = [0] * 7
for i in range(1, 7):
    percentuais[i] = (votos[i] / total_votos) * 100 if total_votos != 0 else 0

print("\nSistema Operacional     Votos   %")
print("-------------------     -----   ---")
opcoes = ["Windows Server", "Unix", "Linux", "Netware", "Mac OS", "Outro"]
for i in range(1, 7):
    print(f"{opcoes[i-1]:<25}{votos[i]:<8}{percentuais[i]:.1f}%")
print("-------------------     -----   ---")
print(f"Total                    {total_votos}")


indice_vencedor = votos.index(max(votos))
vencedor = opcoes[indice_vencedor - 1]
print(f"\nO Sistema Operacional mais votado foi o {vencedor}, com {votos[indice_vencedor]} votos, correspondendo a {percentuais[indice_vencedor]:.1f}% dos votos.")
```

```python
#QUESTÃO 17 (exemplo feito)

num  = int(input("Informe o número: "))
numBinario =[]



if num > 127 or num < -127:
    print("Não é possível representar")

else:
    # Analisa o sinal
    if num < 0 :
        sinal = 1
        num = num * (-1) # Converte o número para um valor positivo - facilitar condição do while

    else:
        sinal = 0


    # Calcula a magnitude
    while num > 0:
            resto = num % 2
            numBinario.append(resto)
            num = num // 2

    # Completa com zero, se necessário. Deve ter um total de 7 bits (magnitude)
    for i in range(7-len(numBinario)):
        numBinario.append(0)

    numBinario.append(sinal) # adiciona o sinal

    #Inverte o número
    numBinario.reverse()
    print(numBinario)
```

```python
#QUESTÃO 27

tamanho = 2

matriz1 = []
matriz2 = []
matriz3 = []

for i in range(tamanho):
    linha1 = []
    for j in range(tamanho):
        numero = int(input("Digite o valor da matriz 1 na posição {},{}: ".format(i, j)))
        linha1.append(numero)
    matriz1.append(linha1)

for i in range(tamanho):
    linha2 = []
    for j in range(tamanho):
        numero = int(input("Digite o valor da matriz 2 na posição {},{}: ".format(i, j)))
        linha2.append(numero)
    matriz2.append(linha2)

print("a) SOMAR AS DUAS MATRIZES \nb) SUBTRAIR A PRIMEIRA COM A SEGUNDA")
print("c) ADICIONAR UMA CONSTANTE (2) NAS DUAS MATRIZES \nd) IMPRIMIR AS MATRIZES")

opcao = input("Digite de acordo com a tabela acima: ")

while opcao not in ["a", "b", "c", "d"]:
    print("OPÇÃO INVÁLIDA! FIQUE ATENTO A TABELA!")
    opcao = input("Digite de acordo com a tabela acima: ")

if opcao == "a":
    for i in range(tamanho):
        linha3 = []
        for j in range(tamanho):
            linha3.append(matriz1[i][j] + matriz2[i][j])
        matriz3.append(linha3)
else:
  if opcao == "b":
    for i in range(tamanho):
        linha3 = []
        for j in range(tamanho):
            linha3.append(matriz1[i][j] - matriz2[i][j])
        matriz3.append(linha3)
  else:
    if opcao == "c":
      for i in range(tamanho):
        for j in range(tamanho):
            matriz1[i][j] += 2
            matriz2[i][j] += 2

    print("MATRIZ 1:")
    for linha1 in matriz1:
        print(linha1)

    print("MATRIZ 2:")
    for linha2 in matriz2:
        print(linha2)

    else:
      if opcao == "d":
        print("MATRIZ 1:")
        for linha1 in matriz1:
          print(linha1)

        print("MATRIZ 2:")
        for linha2 in matriz2:
          print(linha2)

if opcao in ["a", "b"]:
    print("MATRIZ RESULTANTE:")
    for linha3 in matriz3:
        print(linha3)
```

```python
#QUESTÃO 29

estoque = [[0] * 6 for _ in range(3)]  # 3 marcas e 6 tamanhos
tamanhos = [35, 36, 37, 38, 39, 40]
marcas = ["A", "B", "C"]

for i in range(len(marcas)):
    print(f"Insira os dados do estoque para a marca {marcas[i]}:")
    for j in range(len(tamanhos)):
        quantidade = int(input(f"Quantidade para o tamanho {tamanhos[j]}: "))
        estoque[i][j] = quantidade


for i in range(len(marcas)):
    marca = estoque[i]
    maior_quantidade = max(marca)
    indice_tamanho = marca.index(maior_quantidade)
    tamanho_correspondente = tamanhos[indice_tamanho]
    print(f"Numeração com maior quantidade em estoque para a marca {marcas[i]}: {tamanho_correspondente} ({maior_quantidade} unidades)")
```
